```python
class Node:
    def __init__(self,val):
        self.val=val
        self.next=None
```

```python
class MyLinkedList(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.head=None   
        self.size=0
        

    def get(self, index):
        """
        Get the value of the index-th node in the linked list. If the index is invalid, return -1.
        :type index: int
        :rtype: int
        """
        ## 逐一尋訪每一個節點
        ## 1.確認linkedlist是否為空，若是空值回傳 -1
        ## 2.檢查index值是否有超過
        ## 3.回傳尋訪的每一個value (cur.val)
        
        if self.head == None:
            return -1
        if index<0 or index>=self.size:
            return -1
        
        cur = self.head        
        
        for i in range(index):
            cur=cur.next
            print(cur.val)
        return cur.val
        
        
        

    def addAtHead(self, val):
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        :type val: int
        :rtype: None
        """
        
        ## 1.建立new node(帶有欲新增的值)
        ## 2.將new node指向原linkedlist的第一個node
        ## 3.把head改成new node
        ## 4.增加size
        
        new_node=Node(val)
        new_node.next=self.head  ##原linkedlist的第一個node
        self.head=new_node
        
        self.size+=1
        

    def addAtTail(self, val):
        """
        Append a node of value val to the last element of the linked list.
        :type val: int
        :rtype: None
        """
        
        ## 1.建立new node(帶有欲新增的值)
        ## 2.把原linkedlist指向指向new node
        ## 3.如果原linkedlist是空的，self.head需要定義
        ## 4.增加size
        
        if self.head==None:
            self.head=Node(val)
        
        else:  
            cur=self.head         ## 確認current node在第一個
            while cur.next:
                cur=cur.next
                
            new_node=Node(val)    ## 建立new node
            cur.next=new_node     ## 原linkedlist指向指向new node
            
        self.size+=1
        
        

    def addAtIndex(self, index, val):
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        :type index: int
        :type val: int
        :rtype: None
        """
        ## 如果index=size，則將值加在最尾巴
        ## 1.建立new node(帶有欲新增的值)
        ## 2.原linkedlisyt走訪到 index-1 都不做改變
        ## 3.將新建立new node內的pointer指向原linklist的cur.next
        ## 4.原linklist的cur.next走到new node
        ## 5.檢查index值
        
        
        if index==self.size:
            self.addAtTail(val)
            return
        
        elif index < 0 or index >=self.size:
            return
        
        else:
            cur=self.head           ## 確認current node在第一個
            new_node=Node(val)
            
            for i in range(index-1):
                cur=cur.next
                 
            new_node.next=cur.next   ## 將新建立new node內的pointer指向原linklist的cur.next
            cur.next=new_node        ## 原linklist的cur.next走到new node
            
            self.size+=1
             
        
               

    def deleteAtIndex(self, index):
        """
        Delete the index-th node in the linked list, if the index is valid.
        :type index: int
        :rtype: None
        """
        
        ## 1.原linkedlisyt走訪到 index-1 都不做改變
        ## 2.原linklist的cur.next走到cur.next.next
        ## 3.檢查index值
        
        
        if self.head == None:
            return -1
        
        elif index<0 or index>=self.size:
            return -1
        
        else:
            cur=self.head
            for i in range(index-1):
                cur=cur.next
                
            cur.next=cur.next.next    ## 原linklist的cur.next走到cur.next.next
            
            self.size-=1      
        
    
```
